__init__ method also called initializer or constructor is a method that automatically calls each time an instance / object is created.
A self parameter is essential because it is a reference to specific instance of class being created.It allows to access and modify attributes and methods unique to that specific instance ,distinguishing from other instances.
When multiple instances are created from class ,self ensures that each object maintain its own data.

```python
class Employee:

  def __init__(self,first,last,pay):  #constructor that initialize each time a object is created
    self.first=first    #self.varaible=arguement
    self.last=last
    self.pay=pay
    self.email=first+'.'+last+'@org.com'

  def fname(self):   #self is a reference to current instance of class
    """This method returns the full name of Enployee """

    fullname=self.first +' '+self.last
    return fullname

```

#Class Variables

```python
#class variable is special type ofe variable that is shared by all the instances

#           Class Variable                	               Instance Variable
#Scope	    Shared by all objects	                         Unique to each individual object
#Defined	  Inside the class body (outside methods)	       Inside methods (typically __init__, using self in Python)
#Usage	    Constants, shared configurations, counters	   Data unique to a specific object (e.g., a person's name or age)
```

```python
class Employee:

  """ This is the Employee class """
  no_of_employees=0
  raise_amount=1.03 #3 percent salary raiase

  def __init__(self,first,last,pay):
    self.first=first
    self.last=last
    self.pay=pay

    Employee.no_of_employees +=1

  def apply_raise(self):
    """ This function raises the pay of employee based on raise_amount variable"""
    self.pay= int(self.pay * self.raise_amount)  #always use class name or reference(self) with class varialbe
```

why class name or reference (self) is to be used to access the class variable?

Python follows the LEGB rule (Local, Enclosing, Global, Built-in) for variable lookup.

When you use a bare name (e.g., x), Python looks for it first in the local scope of the method.
Class variables exist in the class namespace, which is not automatically searched during standard method execution.
Without self. or ClassName., Python treats the name as a local variable and will raise a NameError if it hasn't been defined within that specific method.

```python
emp1=Employee('Ali','hamza',10000)
emp2=Employee('ahmed','baryal',15000)
```

```python
print(emp1.pay)
emp1.apply_raise()
print(emp1.pay)

print(emp2.__dict__)   #no raise amount varaible here
```

#Methods
Regular methods , class methods , Static methods


Regular method are used for operations that need to know about specific object's data .

* Access: can access both instance level(self.name) and class level(self.class_variable)
* First parameter : must take self (naming convention ) as first parameter which points to current instance
* Call : Usually call on an object.



```python
class Employee:

  """ This is Employee class"""

  raise_amount=1.04

  def __init__(self,name,pay):               #constructor or initializer
    self.name=name
    self.pay=pay

  def pay_raise(self):                       #Regular method
    self.pay=int(self.pay * self.raise_amount)

```

Class methods are used when we need to access and modify the data that belongs to the class as a whole rather than specific object

* Access: can access and modify class level data ; they can not access instance specific attribute
* First parameter: must be cls(naming convention can be other as well)
* decorator: must be marked as class level decorator
* Common Use: factory methods - an alterantive way to create objects(e.g create a user object from JSON string instead of standard parameters)

```python
class Employee:

  """ This is Employee class"""

  raise_amount=1.04

  def __init__(self,name,pay):               #constructor or initializer
    self.name=name
    self.pay=pay
    self.email=name+'@org.com'

  def pay_raise(self):                       #Regular method
    self.pay=int(self.pay * self.raise_amount)

  @classmethod
  def set_raise_amt(cls,amount):
    cls.raise_amount=amount

  @classmethod
  def from_string(cls, emp_str):        #method start with 'from' (naming convention) is usually alterantive for constructor
    name,pay=emp_str.split('-')
    return cls(name,pay)
```

Static method
* static methods are essentialy regular functions that live inside the class for organzational reason or logical connection with class.They donot depend on class or object state.
* Access: can not access instance(self) or class(cls) data directly
* first parameter: doses not take an implicit first parameter
* decorator: must be marked as @staticdecorator
* Common Use: Utility Functions--tasks related to class domain but independent of its data (e.g math calculation or data validation)

```python
class Employee:

  """ This is Employee class"""

  raise_amount=1.04

  def __init__(self,name,pay):               #constructor or initializer
    self.name=name
    self.pay=pay
    self.email=name+'@org.com'

  def pay_raise(self):                       #Regular method
    self.pay=int(self.pay * self.raise_amount)

  @classmethod
  def set_raise_amt(cls,amount):
    cls.raise_amount=amount

  @classmethod
  def from_string(cls, emp_str):        #method start with 'from' (naming convention) is usually alterantive for constructor
    name,pay=emp_str.split('-')
    return cls(name,pay)

  @staticmethod
  def is_workday(day):
      if day.weekday() == 5 or day.weekday() == 6:
        return False
      return True
```
