#Tuple


*   Immutable orderd collection of elements
*   It can hold different data type elements
*   Tuple are heterogeneous, ordered and immutable





###Creating a Tuple

```python
#A tuple is created using ()
a=(1,'b',2.5)
#Using built in function
b=tuple((4,6,8))
print(b)
print(a)
```

###Basing Operations of Tuple


*   Accessing of elements
*   Concatenation of Tuples
*   Slicing of Tuple
*   Deleting a Tuple

```python
#Elements can be accessed using index as in the list
a=(2,4,6,8)
a[2]
```

```python
#Tuple unpacking
x,y,z,t=a
print(x)
print(y)
print(z)
```

```python
#Concatenation of Tuple
#using + operator
#Olny same data type can be concatenated means list and tuple can not be concatenated
tup1=(1,2,3)
tup2=('a','b','c')
tup3=tup1 + tup2
print(tup3)
```

```python
#Slicing a Tuple
#means creating a new tuple from a subset of existing tuple
#[start:stop:step]
a=(1,2,3,4,5,6,7,8)
print(a)
b=a[0:5:2]
print(b)
```

```python
#Reversing a tuple
print(b[::-1])
```

```python
t=(1,2,3,4,5,6,7,8,9)
x=t[::-2]
print(x)
```

```python
#Deleting a tuple
#because tuple are immutable we can not delete element inside it ,instead we can delete whole tuple
tup=(2,6,9,13)
del tup
```

```python
#Tuple unpacking with *
#* is used to gather multiple items from a tuple
# a scenario is that when we only want specific elements from tuple and rest together.
a,*b,c=t
print(a,b,c)
```
