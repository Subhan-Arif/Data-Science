#Dictionary

Stores value in key ,value pair
Keys are:

*   case sensitive
*   must be immutable
*   duplication of keys are not allowed



```python
#Creating a Dictionary
x={1:"a",2:"b",3:"c"}
print(x)

#using dict constructor
t=dict(a=1,b=2,c=3)
print(t)
```

```python
#Accessing dictionary
#using key
d={"name":"ali","age":"22","gender":"male"}
print(d["name"])

#using get function
print(d.get("age"))
```

```python
#Adding and updating
d={1:"a",2:"b",3:"c"}
d[4]="d"
print(d)

#updaing existing one
d[1]="A"
print(d)
```

###Removing Dictionary Item


*   del : removes item by key
*   pop() : remove an item by key and returns its value.
*   clear() :Empities the dictionary
*   popitem(): Removes and returns the last key-value pair



```python
d1={1:"name",2:"age",3:"gender"}

#using del
del d1[1]
print(d1)

#using pop()
val=d1.pop(2)
print(val)

#using clear
d1.clear()
print(d1)

d1[4]="d"
d1[5]="e"
d1[6]='f'
print(d1)
#using popitem()
d1.popitem()
```

```python
#iterating through a dictionary
d2={"name":"ahmad","age":"21","gender":"male"}

#iterate over the key
for key in d2:
  print(key)
```

```python
#iterate over the value
for value in d2.values():
  print(value)
```

```python
#iterate over key and value
for key ,value in d2.items():
  print((key) + " is " + (value))
```
