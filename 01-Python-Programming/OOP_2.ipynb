#Inheritance

If the class we want to create is a specilaized version of existing class,we can use inheritance. When one class inherits from another, it automatically takes on all the attributes and methods of the first class. The original class is known as parent class and new calss is called the child class. The child class inherit every attribute and method from its parent class but is also free to define new attribute and method from its parent class but is also free to define new attributes and methods of its own

```python
class Employee:
  """ This is EMployee class for Organization"""

  raise_amount=1.05

  def __init__ (self,first,last,pay):
    self.first=first
    self.last=last
    self.pay=pay
    self.email=first + last + ' @company.org'


  def fullname(self):
    return (self.first + self.last)

  def apply_raise(self):
    self.pay=int(self.pay *self.raise_amount)

```

Super() is built in function that returns a proxy object allowing a subclass to call method from its parent class(superclass)

we can also use parent class name instead of super() ,but that leads to dupliacte initialization in multiple inheritance

In complex scenario where more than one parent class super does not look immediate parent class instead it follows MRO(method resolution order) to ensure every class in initialized exactly once

```python
#developer class
class Developer(Employee):
  raise_amount=1.10  #10 percent for developer and rest all emlpyee have same raise amount as in the parent class

  def __init__(self,first,last,pay,prog_lang):   # we are able to customize the existing class
    super().__init__(first,last,pay)
    self.prog_lang=prog_lang
```

```python
class Manager(Employee):
  """ This is child class of Employee"""

  def __init__(self,first,last,pay,employees=None):   #dont pass mutabel data type as default arguments
    super().__init__(first,last,pay)
    if employees is None:
      self.employees =[]
    else:
      self.employees=employees

  #add employee
  def add_employee(self,emp):
    if emp not in self.employees:
      self.employees.append(emp)

  #remove employe
  def remove_employee(self,emp):
    if emp in self.employees:
      self.employees.remove(emp)


  #print all employees
  def print_emps(self):
    for emp in self.employees:
      print("--", emp.fullname())
```

```python
manager1=Manager('sue','smith',90000,[dev_1])
```

```python
print(manager1.email)
manager1.print_emps()
```

```python
manager1.add_employee(dev_2)
manager1.print_emps()
```

```python
manager1.remove_employee(dev_1)
manager1.print_emps()
```

#Special (Magic / Dunder) Method

A method that starts and end with __ are called dunder method that are not  called directly but automatically by Pythons built in operator and function to determine how object of class behave


__repr__

* Developers representation return an unambigious string for debugging; ideally reconstruct the object

__str__

* User representation: return a user freindly readable string for end users


Always implement __repr__: It is the global fallback for string representation. If __str__ is missing, Python uses __repr__ instead.

Safety: When implementing __repr__, avoid including sensitive data like passwords to prevent accidental exposure in logs.

Consistency: Use __format__ if you need custom f-string support or machine-stable representations for logging.

```python
class Employee:
  """ This is EMployee class for Organization"""

  raise_amount=1.05

  def __init__ (self,first,last,pay):
    self.first=first
    self.last=last
    self.pay=pay
    self.email=first + last +' @company.org'


  def fullname(self):
    return (self.first + self.last)

  def apply_raise(self):
    self.pay=int(self.pay *self.raise_amount)

  def __repr__(self):
    return "Employee('{} ,{} ,{})".format(self.first ,self.last ,self.pay)

  def __str__(self):
    return '{} - {}'.format(self.fullname(), self.email)


  def __len__(self):
    return len(self.fullname())
```

#Property Decorators

it allows to treat a method as an attribute. It is used to define managed attribute giving control how value is accessed, set or delete with out changing the syntax used to interact with the object

```python
class Employee:
  """ This is EMployee class for Organization"""

  raise_amount=1.05

  def __init__ (self,first,last,pay):
    self.first=first
    self.last=last
    self.pay=pay
    self.email=first + last +' @company.org'


  def fullname(self):
    return (self.first + self.last)

```

```python
emp_1.first='Ahmed'
```

```python
print(emp_1.first)
print(emp_1.email)
print(emp_1.fullname())
```

As we see that changing the first name is not changing the email automatically ,here comes the property decorator that helps do that


Usually, if you want to change how a variable is handled, you have to write "getter" and "setter" methods (e.g., get_email() and set_email()). This breaks existing code that uses obj.email.
The @property decorator allows you to keep the simple obj.email syntax while running custom logic behind the scenes.


```python
class Employee:
  """ This is EMployee class for Organization"""

  raise_amount=1.05

  def __init__ (self,first,last,pay):
    self.first=first
    self.last=last
    self.pay=pay


  @property
  def email(self):
    return '{}.{}@gmail.com'.format(self.first,self.last)

  @property
  def fullname(self):
    return (self.first + self.last)

  @fullname.setter
  def fullname(self,name):
    first, last =name.split(' ')
    self.first=first
    self.last=last


  @fullname.deleter
  def fullname(self):
    print("Delete Name!")
    self.first=None
    self.last=None
```

```python
emp_1=Employee('ali','Hamza',1000)
emp_1.first='Ahmed'
print(emp_1.first)
print(emp_1.email)
print(emp_1.fullname)
```

```python
#The above is getter
#now if i want to set fisrt and last name from full name we can do this using setter
emp_1.fullname="Subhan Arif"
print(emp_1.first)
print(emp_1.email)
print(emp_1.fullname)
```
