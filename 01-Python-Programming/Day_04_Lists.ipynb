###Adding Element in to List

*   append() adds an element at the end of list
*   extend() adds multiple elements at the end of list
*   insert() adds an element at the specific postion
*   clear() removes all items of list



```python
a=[]

a.append(2)
print(a)

a.extend([3,4,5])
print(a)

a.insert(3,9)
print(a)

a.clear()
print(a)
```

###Updating item of the list

```python
a=[1,2,3,4,5]
a[2]=4 #it updates the item at index 2
print(a)
```

###Removing item / element from the list


*   remove() it remove the first occurence of element / item
*   pop() it removes the element at desired index or the last index element if not specified
*   del statement() Deletes an element at specified index




*   The del statement is used to delete names (references) from a scope (namespace). It doesn't necessarily delete the underlying object from memory immediately (Python's garbage collector handles that when the object is no longer referenced anywhere), but it makes the variable name undefined in the current scope.


```python
a=[1,2,1,3,4,5,6]
a.remove(1)

print(a)

a.pop(0)
print(a)

a.pop()
print(a)

del a[2]
print(a)
```

###List comprehension

```python
#It is a way to create a list using a single line of code
#[ expression for item in iterable ]
#with if
#[ expression for item in iterable if condition ]
cube_list=[x**3 for x in range(10)]
print(cube_list)
```

```python
#list comprehension executes in sequence exactly like an equivalent standard for loop structure.
#The processing order flows from left to right
#,with outer loops defined first, followed by inner loops and conditional filters.
coords = [(x, y) for x in [1, 2] for y in [10, 20]]
```

Execution sequence:

    x gets the value 1.
    Inner loop starts:
        y gets 10. Expression (x, y) evaluates to (1, 10), appended to coords.
        y gets 20. Expression (x, y) evaluates to (1, 20), appended to coords.
    x gets the value 2.
    Inner loop starts again:
        y gets 10. Expression (x, y) evaluates to (2, 10), appended to coords.
        y gets 20. Expression (x, y) evaluates to (2, 20), appended to coords

In python a list does not stores the actual value directly instead it stores reference (pointers) to objects in memory, thats why list can store different type of items and can be able to update them (mutable)

####alising


```python
#because list are mutable and when we assign one list variable to another variable ,both points toward the same object in memory.This is called aliasing.
#changes made through one variable can be seen through other as well
a=[1,2,3,4]
b=a
print(a)
print(b)
```

```python
b.append(4)
print(b)
print(a)
```
